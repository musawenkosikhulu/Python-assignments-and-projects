```python
import matplotlib.pyplot as plt          #This code import a library matplotlib so that we can use it to visualise our graphs
import numpy as np                       #This code will a library numpy which is used for mathematical functions like pi, sin, cos etc


x=np.linspace(-6,6)                      #This chuck of code for the interval [-6,6] on the grid

color_change = ["c","y","k","b","g","m"] #This chuck of code is a list of different color of the graphs I want plot using different colors so the alphabets represent different colors
Title_change = ["$f_{t=-3}~(x)$","$f_{t=-2}~(x)$","$f_{t=-1}~(x)$","$f_{t=0}~(x)$","$f_{t=1}~(x)$","$f_{t=2}~(x)$"] #This we will excute our graph for different values of t so chuck gives the title of our graphs for diffirent t values

for t in range(-3,3):                    #This loop is for different type of t from -3 to 2 and we know the range fucntion is exclusive, it won`t excute the last part
    y=np.exp(-1*(x-3*t)**2) *np.sin((np.pi*3)*(x-t)) #This is our given fuction where we change different value of t which taken of by the loop
    
    plt.figure                           #This code will show us our different figures 
    plt.subplot(2,3,t+4)                 #Since we have to code subplots, we must introduce the variable t but since our t does not start from 1 we had to plus it by 4
    plt.plot(x,y, color_change[t])       #This chuck of code plots our graph and changes the colors of the graph from our defined color list on the top and it taking the advantage of the loop when it iterate it updates the color
    plt.grid()
    plt.xlabel("$x-axis$")               #This part just sets x - axis as the for loop iterates
    
    plt.ylabel("$y-axis$")               #This part just sets y - axis as the for loop iterates
    plt.title(Title_change[t+3])

    plt.legend(["$f_t(x) = e^{-(x-3t)^{2}}sin(3\pi(x-t))$"]) #This set our legends so that we see what graph are we plotting
    plt.rcParams["figure.figsize"]=[25,20]
    fig = plt.gcf()                      #allows you to get a reference to the current figure
    fig.set_size_inches(20, 15)          #This will give us the size inches of the plot
    fig.savefig('Musawenkosi plots.png', dpi=100)# This code saves my figure to the given name
    plt.subplots_adjust(wspace=0.3,hspace=0.5) #This fixes the spaces in between the plots.
```

```python
import random       #This chuck of code here imports the library random so that we can use it to N random points
import numpy as np  #Which are uniformly distributed
                    #The numpy library is used to generated convient mathematical functions  like sin,cos,exp and pi             
   
    
N=int(input('Enter the values: ')) #This code ask the user to enter a large number N and assign it to an integer
radius=1                           #we choose the radius to be 1 because place the circle in the origin the 

count_p_inside=0
x_p = np.random.uniform(0,2,(N))   #This code generates the N random points (x_i)
y_p = np.random.uniform(0,2,(N))   #This code generates the N random points (x_i)

for i in range(N):                 #In order to repeat N number of times we introduce a four loop which repeat the process N times
    x=x_p[i]                       #The randomly generated x_i codes are repeated in accordance with the for loop
    y=y_p[i]                       #The same thing will happen with y where we randomly generated y_i codes are repeated in accordance with the for loop
    
    if (x-1)**2+(y-1)**2<=radius**2: #In order to get the point inside the circle we have to restrict the points using a simple circle equation x**2+y**2 = r**2
        count_p_inside+=1            #So the points should satisfy that if the randomly generated (x_i,y_i) points should not be greater than radius r
print("Number of points inside the circle:",count_p_inside)#counting the numbers of  points falls inside the circle.
pi=4*count_p_inside/N                #This estimate pi using the given number of points inside the circle and N number of times
print('estimate value of pi: ',pi)            #This chuck code prints out the estimated pi
```

### ***(b)*** 

The (b) part where we count the number of times for points inside the circle is covered by **count_p_inside**

### ***(c)***

The basic definition of probability of event to happen states that $ P(E) = \frac{Number~of~favourable ~outcomes}{Total~Number~of~outcomes} $, So in our case we wamt to count points inside the circle which can be represented by $A_c$. The cricle is inside the square therefore the total number of point are contained inside the square which can be represented by an area of $A_s$. Hence, we get the equation $p = \frac{A_c}{A_s}$.

### ***(d)***

We know that the probability is given by $p = \frac{A_c}{A_s}$ we know that $A_c = \pi r^2$ and $A_s = 2r \times 2r = 4r^2$  \
$\implies that ~ p = \frac{A_c}{A_s} = \frac{\pi r^2}{4r^2}$ \
$p = \frac{\pi}{4}$ $\blacksquare$


### ***(e)***

We can get or rather estimate the probability of the points falling inside circle by taking the total number of point(inside the circle) $n$ counts divided by the large number $N$ number of iteration. \
$p = \frac{n}{N} \equiv p = \frac{\pi}{4}$ \
$\implies \pi = \frac{4n}{N}$ from simple mathematical manipulation. $\blacksquare$

### ***(f)*** 

from the code in **(a)** we can estimate $\pi$ by taking **count_p_inside** and **N**, thenafter, we use the equation $\pi = \frac{4n}{N}$
